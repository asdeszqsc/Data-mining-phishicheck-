<h4> hyperparameter 구하기</h4> <h4>EDA한 시각화 자료와 통계량 등을 이용하여 해석</h4>

```python
x_train, x_test, y_train, y_test = train_test_split(df, df_target, test_size=0.1, random_state=38)
```

<h2>(1) KNN <h2/>

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
knn_euc_kfold_accuracy_list = []
knn_man_kfold_accuracy_list = []

knn_man_accuracy_list = []
knn_euc_accuracy_list = []

for k in range(3, 21):
    #k=3 부터 20까지 knn을 metric이 manhattan로 선언 
    knn1 = KNeighborsClassifier(n_neighbors=k, metric = 'manhattan')
    
    #k=3 부터 20까지 knn을 metric이 euclidean로 선언 
    knn2 = KNeighborsClassifier(n_neighbors=k, metric = 'euclidean')
    
    
    scores_man_kfold_i = cross_val_score(knn1,x_train, y_train.values.ravel(), cv=10, scoring='accuracy')
    #knn 알고리즘을 k값에 따라 10 fold cross - validation 수행
    
    scores_euc_kfold_i = cross_val_score(knn2,x_train, y_train.values.ravel(), cv=10, scoring='accuracy')
    #knn 알고리즘을 k값에 따라 10 fold cross - validation 수행

    
    knn_man_accuracy_list.append(scores_man_kfold_i.mean())          
    #10 fold cross validation 수행 후 k 값에따른 fold 별 정확도의 평균을 list에 저장 
    
    knn_euc_accuracy_list.append(scores_euc_kfold_i.mean())          
    #loo cross validation 수행 후 k 값에따른 fold 별 정확도의 평균을 list에 저장 
       
        
knn_euc_kfold_scores = pd.Series(knn_euc_kfold_accuracy_list)
knn_man_scores = pd.Series(knn_man_accuracy_list)
#list를 Series로 변환
```

```python
best_score = 0
best_k = 0
for i in range(0,18):
    if(knn_man_scores[i] > best_score):
        best_score = knn_man_scores[i]
        best_k=i
```

```python
print("best_score   best_k")
print(best_score, best_k)
```

<h2>(2) Logistic Regression <h2/>

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
import warnings
warnings.filterwarnings(action='ignore')
#warnings 메시지 무시

model = LogisticRegression()
grid={"C":np.logspace(-3,3,7), "penalty":["l1","l2"]}
model_cv = GridSearchCV(model, grid, cv=10)
model_cv.fit(x_train, y_train)
print("best parameters : ",model_cv.best_params_)
print("best cross-validation : ", model_cv.best_score_)
print("정확도 : ", model_cv.score(x_test,y_test))
```

```python
model_test = LogisticRegression(C = 10.0, penalty = "l2")
model_test.fit(x_train, y_train)
prediction = model_test.predict(x_test)
print(classification_report(prediction,y_test))
```

```python
df = pd.DataFrame(df, columns = ['Prefix_suffix','having_Sub_Domain','Alexa_Ranking', 'Count_Redirection','External_Load_Script','getAnchorResult','form_action_Handler','getLinksInTags','Domain_registration_length','Shortening_Service',"favicon","URL_Length"])
x_train, x_test, y_train, y_test = train_test_split(df, df_target, test_size = 0.1,random_state = 38)
model2 = LogisticRegression()
grid2={"C":np.logspace(-3,3,7), "penalty":["l1","l2"]}
model_cv2 = GridSearchCV(model2, grid2, cv=10)
model_cv2.fit(x_train, y_train)
print("best parameters : ",model_cv2.best_params_)
print("best cross-validation : ", model_cv2.best_score_)
print("정확도 : ", model_cv2.score(x_test,y_test))
```

```python
model_test = LogisticRegression(C = 1.0, penalty = "l2")
model_test.fit(x_train, y_train)
prediction = model_test.predict(x_test)
print(classification_report(prediction,y_test))
```

<h2>(3) Random Forest<h2/>

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

rf_model = RandomForestClassifier(random_state = 0)
param_grid = {
    'n_estimators' : range(50, 650, 100),
    'max_depth' : [6, 8, 10, 12]}
```

```python
print("best parameters : ",rf_grid.best_params_)
print("best cross-validation : ", rf_grid.best_score_)
print("정확도 : ", rf_grid.score(x_test,y_test))
```

<h2>(4) SVM linear <h2/>

```python
from sklearn.svm import SVC # SVM classifier 사용위한 import

#hyperparameter는 슬랙변수 가중치 C 이므로 c를 변화주면서 hyperparameter를 tuning한다
slacklist = np.arange(0.1,10.1,0.1) #0~10 까지 slacklist생성
```

```python
#Grid Search
from sklearn.model_selection import GridSearchCV 
  
# parameter 정의
param_grid = {'C': slacklist,  
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
              'kernel': ['linear']}  
#SVM grid Search  
SVC_linear_grid = GridSearchCV(SVC(), param_grid, refit = True, n_jobs=4 ,verbose = 3).fit(x_train,y_train.values.ravel())
```

```python
#model evaluation
print("test set score : {}".format(SVC_linear_grid.score(x_test,y_test)))

#best parameters and best score
print("best parameters : {}".format(SVC_linear_grid.best_params_))
print("best score : {}".format(SVC_linear_grid.best_score_))
```

```python
best_linear_clf = SVC(kernel = 'linear', C = 8.5, gamma = 1).fit(x_train,y_train)
#best parameter : {'C': 8.5, 'gamma': 1, 'kernel': 'linear'} 를 이용해서 SVM_linear classifier 생성

test_predict_linear_SVM = best_linear_clf.predict(x_test)


print("linear SVM classifier\n")
print(classification_report(y_test, test_predict_linear_SVM,target_names=['class','class 1'] ))
#linear SVM의 정밀도, 재현율, F1 점수를 나타낸다.
```
