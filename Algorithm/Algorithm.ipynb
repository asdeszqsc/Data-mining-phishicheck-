<h2>(4) SVM linear <h2/>

```python
from sklearn.svm import SVC # SVM classifier 사용위한 import

#hyperparameter는 슬랙변수 가중치 C 이므로 c를 변화주면서 hyperparameter를 tuning한다
slacklist = np.arange(0.1,10.1,0.1) #0~10 까지 slacklist생성
```

```python
from sklearn.model_selection import GridSearchCV 
  
# parameter 정의
param_grid = {'C': slacklist,  
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
              'kernel': ['linear']}  
#SVM grid Search  
SVC_linear_grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3).fit(x_train,y_train.values.ravel())
```

```python
#model evaluation
print("test set score : {}".format(SVC_linear_grid.score(x_test,y_test)))

#best parameters and best score
print("best parameters : {}".format(SVC_linear_grid.best_params_))
print("best score : {}".format(SVC_linear_grid.best_score_))
```

```python
best_linear_clf = SVC(kernel = 'linear', C = 8.5, gamma = 1).fit(x_train,y_train)
#best parameter : {'C': 8.5, 'gamma': 1, 'kernel': 'linear'} 를 이용해서 SVM_linear classifier 생성

test_predict_linear_SVM = best_linear_clf.predict(x_test)


print("linear SVM classifier\n")
print(classification_report(y_test, test_predict_linear_SVM,target_names=['class','class 1'] ))
#linear SVM의 정밀도, 재현율, F1 점수를 나타낸다.
```

<h2>(5) SVM Kernel <h2/>

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import LinearSVC, SVC
slacklist = np.arange(0.5,10.0,0.5)
param_grid = {'C': slacklist,  
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
              'kernel': ['rbf', 'sigmoid']}  
```

```python
SVC_kernel_result = GridSearchCV(SVC(), param_grid, n_jobs = -1, verbose = 3).fit(x_train,y_train.values.ravel())
```

```python
print("test set score : {}".format(SVC_kernel_result.score(x_test,y_test)))
print("best parameters : {}".format(SVC_kernel_result.best_params_))
print("best score : {}".format(SVC_kernel_result.best_score_))
print(classification_report(SVC_kernel_result.predict(x_test),y_test.values.ravel()))
```

```python
slacklist = np.arange(0.1,10.1,1.0)
param_grid = {'C': slacklist, 
              'gamma': [1, 0.1, 0.01, 0.001], 
              'kernel': ['poly']}
SVC_poly_result = GridSearchCV(SVC(), param_grid, n_jobs = -1, verbose = 3, cv = 3).fit(x_train,y_train.values.ravel())
```

```python
print("test set score : {}".format(SVC_poly_result.score(x_test,y_test)))
print("best parameters : {}".format(SVC_poly_result.best_params_))
print("best score : {}".format(SVC_poly_result.best_score_))
print(classification_report(SVC_poly_result.predict(x_test),y_test.values.ravel()))
```
