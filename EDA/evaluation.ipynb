```python
import pandas as pd #pandas  
import numpy as np # numpy
from sklearn.model_selection import train_test_split #train_test_split 함수 사용
from sklearn.metrics import accuracy_score #정확도 score 사용
import matplotlib as mpl           #mpl           
import matplotlib.pyplot as plt    #plt
```

```python
df = pd.read_csv("../Algorithm/DataSet.csv") #feature 있는 dataset 
df
```

```python
df.info()
#feature 정보 확인
```

```python
print(df['Length_of_Source'][0])
for i in range(0, 15788):
    if((not df['Length_of_Source'][i]) and (not df['duplicated_Tag'][i]) and (not df['Disabling_Right_Click'][i]) and (not df['External_Load_Script'][i]) and (not df['Count_Redirection'][i]) and (not df['getAnchorResult'][i]) and (not df['form_action_Handler'][i]) and (not df['getLinksInTags'][i]) and (not df['onMouseOver'][i]) and (not df['Prompt_in_Popup'][i]) and (not df['DomainName_in_Source'][i]) and (not df['Shortening_Service'][i]) and (not df['favicon'][i]) and (df['label'][i])):
        df['label'][i] = None
    
df = df.dropna()
```

```python
df_target = pd.DataFrame(df['label'], columns = ['label'])
df_target
#target Data 지정
```

```python
x_train, x_test, y_train, y_test = train_test_split(df, df_target, test_size=0.1, random_state=38)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score

#선택된 hyperparameter k=12 , metric= manhattan 사용
knn1 = KNeighborsClassifier(n_neighbors=12, metric = 'manhattan')    
knn1.fit(x_train, y_train)
scores_man_kfold = cross_val_score(knn1,x_train, y_train.values.ravel(), cv=10, scoring='accuracy')         
#10 fold cross validation 수행 후 k 값에따른 fold 별 정확도의 평균을 저장 
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
import warnings

model_test = LogisticRegression(C = 1.0, penalty = "l2")
model_test.fit(x_train, y_train)
prediction = model_test.predict(x_test)
print(classification_report(prediction,y_test))
```

```python
model_test = LogisticRegression(C = 10.0, penalty = "l2")
model_test.fit(x_train, y_train)
prediction = model_test.predict(x_test)
print(classification_report(prediction,y_test))
```
