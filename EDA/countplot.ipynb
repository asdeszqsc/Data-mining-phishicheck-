```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("Features_.csv", encoding='CP949')
```

```python
for feature in data.columns[2:9]:
    ax = sns.countplot(x = feature, data=data, hue='label')
    for p in ax.patches:
        left, bottom, width, height = p.get_bbox().bounds
        ax.annotate("%.0f"%(height), (left+width/2, height*1.01), ha='center')
    plt.show()

for feature in data.columns[10:24]:
    ax = sns.countplot(x = feature, data=data, hue='label')
    for p in ax.patches:
        left, bottom, width, height = p.get_bbox().bounds
        ax.annotate("%.0f"%(height), (left+width/2, height*1.01), ha='center')
    plt.show()
```
